```python
import time
# 開始測量
start = time.time()

#列印出文章標題及網址，並存取圖片
import requests
import json
# to save image by request.urlretrieve
import os
from urllib import request

hders={'User-Agent' :'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'}

res_path = r'./dcard_photography'
if not os.path.exists(res_path):
    os.mkdir(res_path)

#network>filter搜尋posts?    #注意limit數
url="https://www.dcard.tw/service/api/v2/forums/photography/posts?limit=30&before=237655995"

for i in range(0, 2):
    try:
        res = requests.get(url, headers=hders)
        json_string = str(res.text)
        js = json.loads(json_string)

        last_id = js[len(js) - 1]['id']

        for each_article in js:
            print(each_article['title'])
            print('https://www.dcard.tw/f/photography/p/' + str(each_article['id']))
            for n, img_url in enumerate(each_article['mediaMeta']):
                tmp_img_url = img_url['url']
                location = os.path.join(res_path + '/%s_%s.jpg'%(each_article['title'].replace('/',''), n))#n是enumerate自動編號
                print(('\t' + tmp_img_url), end = '')#前面縮排tab，結尾空一格
                request.urlretrieve(tmp_img_url, location) #urlretrieve
                print('\tDone.') #\t是tab的意思
            print()
    except:
        pass
    continue
# 結束測量
end = time.time()
# 輸出結果
print()
print("執行時間：%f 秒" % (end - start))
```
